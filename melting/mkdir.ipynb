```python
import matplotlib.pyplot as plt
import numpy as np
import os
from math import sqrt
```

```python
len=0.9
sqrt32=sqrt(3)/2*len
devi21 = 1/2 *len

num=7
basenum=3
xlen=basenum*num
ylen=basenum*num-6

x00=6
y00=6
x0=x00
y0=y00

b=[]
a=[]

#第一列
i=0
x=x0
y=y0
z=1
while i<ylen-1:
    if z%3==0:
        z+=1
        y+=len
        continue
    b.append([x,y])
    z+=1
    y+=len
    i+=1



x0 += sqrt32
y0 -= devi21
x=x0
y=y0
i=0
z=0
k=0
while i<(xlen-2):
    if i%2==0:
       for j in range(int(ylen/2)):
            b.append([x,y])
            a.append([x,y+len])
            b.append([x,y+2*len])
            y+=3*len
    else:
        b.append([x,y])
        for j in range(int(ylen/2)-1):
            b.append([x,y+len])
            a.append([x,y+2*len])
            b.append([x,y+3*len])
            y+=3*len
        b.append([x,y+len])
    x0+=len
    y0-=devi21*((k%2)-0.5)*2
    k+=1
    x=x0
    y=y0
    i+=1

x=x0
y=y0
z=1
i=0
while i<ylen-1:
    if z%3==0:
        z+=1
        y+=len
        continue
    b.append([x,y])
    z+=1
    y+=len
    i+=1

a=np.array(a)
b=np.array(b)
plt.figure(figsize=(10,10))
plt.scatter(a[:,0],a[:,1],c='r')
plt.scatter(b[:,0],b[:,1],c='b')

```

```python
print(a.shape[0])
print(b.shape[0])
print(a.shape[0]+b.shape[0])

with open('./output.txt', 'w') as f:
    np.savetxt(f,a,fmt='%.2f')
    np.savetxt(f,b,fmt='%.2f')
print('Done!')
```
