```python
import numpy as np
import matplotlib.pyplot as plt

R = 9
k=-0.05

# Dense sampling near x~0.1 where y changes rapidly
x1 = np.geomspace(0.8, 1.0, 30000)     # geometric spacing (dense near 0.1)
x2 = np.linspace(1.0, 3.0, 60000)
x  = np.r_[x1[:-1], x2]

y1 = np.where(x < 1, -1 + (1/x)**4 + k*(R*(1/x)**2-1),
                  k*(R*(1/x)**2-1))

k=0.01
y2 = np.where(x < 1, -1 + (1/x)**4 + k*(R*(1/x)**2-1),
                  k*(R*(1/x)**2-1))

fig, ax = plt.subplots(figsize=(6,4), dpi=200)
ax.plot(x, y1, color='k', lw=5,
        antialiased=True,
        solid_joinstyle='miter',   # avoid rounded bulges at tight turns
        solid_capstyle='butt')     # flat line ends

ax.plot(x, y2, color='r', lw=5,
        antialiased=True,
        solid_joinstyle='miter',   # avoid rounded bulges at tight turns
        solid_capstyle='butt')     # flat line ends

# Avoid hard clipping at the axes box
#ax.set_axis_off()
ax.margins(x=0.02, y=0.02)         # tiny padding so the stroke isn't cut

# (Optional) limit extreme y to keep slopes reasonable on screen
# ax.set_ylim(-1, 1)

plt.show()
```
